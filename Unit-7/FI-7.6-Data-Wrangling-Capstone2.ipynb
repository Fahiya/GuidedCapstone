### Data Cleaning

#### Handle the Missing and NA values

```python
df.isnull().sum()
```

#### calculating percentage of missing data by features:

```python
df_percentage = (df.isnull().sum() / len(df)) * 100

f, ax = plt.subplots(figsize=(15, 8))
plt.xticks(rotation='90')
sns.barplot(x=df_percentage.index, y=df_percentage)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
```

##### The above figure shows the features 'PoolQC', 'MiscFeature', 'Alley', 'Fence' have many null values. The following heatmap shows the same thing.

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False)
```

```python
# We are trying to see how many features have missing values by taking a look at the percentage
nas=pd.DataFrame(df.isnull().sum().sort_values(ascending=False)/len(df),columns = ['percent'])
pos = nas['percent'] > 0
nas[pos]
```

###### It looks like PoolQC and a few other  features have null values more than 80 percent so it might be tempting to drop the columns. But we will not be dropping them. We do isnull on the columns that have NaNs in them and create a new dataframe that has only the columns with NaN in it. And then we do .isnull() on each of them and that creates a two column dataframe that has nothing but True and False. That becomes our feature that we add. We can add these new columns back into our original dataframe and call them, say, LotArea_isnull and then we fillna() the original null columns with typical practice-> 0/mean/max/min, etc.

#### Numerical Variables

```python
dfn = df._get_numeric_data() #We also found numeric data info when we did describe()
dfn.head(11)
```

##### Now that we have separated out our Numerical features, time to see how many numercial features have missing values

```python
nas=pd.DataFrame(dfn.isnull().sum().sort_values(ascending=False)/len(dfn),columns = ['percent'])
pos = nas['percent'] > 0
nas[pos]
```

#### We can also check the heat map for dfn DataFrame

```python
sns.heatmap(dfn.isnull(),yticklabels=False,cbar=False)
```

##### It shows only three Numercial features have null values. So, we are adding the _isnull features for all three now per the discussion above

```python
dfn['LotFrontaget_isnull'] = dfn['LotFrontage'].isnull()
dfn['GarageYrBlt_isnull'] = dfn['GarageYrBlt'].isnull()
dfn['MasVnrArea_isnull'] = dfn['MasVnrArea'].isnull()
```

```python
#Creating a list of dfn.columns and adding the three extra columns in the list with append().
#Then checking if the feature numbers increased by three. 
features_added = list(dfn.columns)
features_added.append('LotFrontaget_isnull')
features_added.append('GarageYrBlt_isnull')
features_added.append('MasVnrArea_isnull')
```

```python
dfn.shape # We shoule see three extra columns added to the existing 38 columns. 
```

#### Now replacing the nulls with mean values. Generally, null for float values are replaced by mean, but we are not longer worried about what method we use, since we have added the extra features with isnull, we are just using mean. 

```python
# Creating a new dataframe after null value replacement
dfn_no_null = dfn.fillna(value=dfn.mean())
```
